# K-means clustering

## import library

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from matplotlib import cm
```

## load data

```python
fname_data = 'assignment_11_data.csv'

feature = np.genfromtxt(fname_data, delimiter=',')

x = feature[:,0]
y = feature[:,1]

number_data     = np.size(feature, 0)
number_feature  = np.size(feature, 1)

print('number of data : {}'.format(number_data))
print('number of feature : {}'.format(number_feature))
```

## plot the input data

```python
plt.figure(figsize=(8,8))   
plt.title('input data')

plt.scatter(x, y, color='blue')

plt.tight_layout()
plt.show()
```

## compute distance
- feature : $n \times m$, center : $1 \times m$, distance : $n \times 1$
- $n$ : number of data, $m$ : number of features

```python
def compute_distance(feature, center): 

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    dist = (feature - np.tile(center, (len(feature),1))) **2
    distance = np.sqrt(dist[:,0] + dist[:,1])
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return distance
```

## compute centroid
- feature : $n \times m$, label_feature : $n \times 1$, value_label : $1 \times 1$, centroid : $1 \times m$
- $n$ : number of data, $m$ : number of features

```python
def compute_centroid(feature, label_feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    x1= []
    y1= []
    for i in range(number_data):
        if label_feature[i] == label:
            x1.append(feature[i,0])
            y1.append(feature[i,1])
    x2 = 0 if len(x1)==0 else np.mean(np.asarray(x1))
    y2 = 0 if len(y1)==0 else np.mean(np.asarray(y1))
    centroid = np.array([x2, y2])
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return centroid
```

## compute label
- distance : $n \times k$, label_feature : $n \times 1$
- $n$ : number of data, $k$ : number of clusters

```python
def compute_label(distance):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    list = []
    for i in range(len(distance)):
        k = distance[i].argmin()
        list.append(k)
    label_feature = np.asarray(list)
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return label_feature
```

```python
def compute_loss(feature, label_feature, number_cluster, centroid):
    temp = 0
    data = np.c_[feature,label_feature]
    for k in range(number_cluster):
        temp += sum(compute_distance(data[data[:,2]==k][:,:2], centroid[k]))
    temp /= number_data
    return temp
```

## the number of clusters $K = 2$

```python
number_cluster          = 2
number_iteration        = 50     # you can modify this value
loss_iteration_02       = np.zeros(number_iteration)
centroid_iteration_02   = np.zeros((number_iteration, number_cluster, number_feature))
label_feature_02        = np.random.randint(0, number_cluster, size=(number_data))
```

```python
# ++++++++++++++++++++++++++++++++++++++++++++++++++
# complete the blanks
#
centroid_02 = np.zeros((number_cluster, number_feature))
distance_02 = np.zeros((number_data, number_cluster))
label_cluster_02 = number_cluster
for i in range(number_iteration):
    for k in range(number_cluster):
        centroid_02[k] = compute_centroid(feature, label_feature_02, k)
        distance_02[:,k] = compute_distance(feature, centroid_02[k])
    label_feature_02 = compute_label(distance_02)
    loss_02 = compute_loss(feature, label_feature_02, number_cluster, centroid_02)
    centroid_iteration_02[i] = centroid_02
    loss_iteration_02[i] = loss_02
    #print('[%d] loss : %.5f' % (i, loss_02))

#
# ++++++++++++++++++++++++++++++++++++++++++++++++++
```
